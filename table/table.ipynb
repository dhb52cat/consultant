```python
import openpyxl
# 拆分合并单元格
# 加载工作簿
workbook = openpyxl.load_workbook('押品表结构清单1.xlsx')

sheet_with_names=workbook['Sheet1']
max_row = sheet_with_names.max_row

# 获取 A 列的最大行数
max_row = sheet_with_names.max_row
print(max_row)
# 遍历 A 列的每一行
for row in range(1, max_row + 1):
    # 获取 A 列当前行单元格的值
    sheet_name = sheet_with_names.cell(row=row, column=1).value

    # 检查单元格值是否为空
    if sheet_name is not None:
        # 检查该名称的工作表是否存在于工作簿中
        name=''
        if sheet_name in workbook.sheetnames:
            name=sheet_name
        elif sheet_name.lower() in workbook.sheetnames:
            name=sheet_name.lower()
        if name!='': 
            # 获取对应的工作表
            target_sheet = workbook[name]
            inmaxrow=target_sheet.max_row
            # 获取该工作表第一个单元格的值
            chname = target_sheet.cell(row=1, column=2).value

            for row in range(3,inmaxrow):
                print(f'{target_sheet.cell(row=row,column=1)}')
            print(f'工作表 {sheet_name} 存在，其第一个单元格的值为: {chname}')
        else:
            print(f'工作表 {sheet_name} 不存在。')
    else:
        print(f'A 列第 {row} 行的值为空。')

# 关闭工作簿
workbook.close()
```

```python
import openpyxl
import re

# 加载工作簿
workbook = openpyxl.load_workbook('押品表结构清单1.xlsx')

# 获取当前活动工作表
source_sheet = workbook['Sheet1']

# 获取 A 列的最大行数
max_row = source_sheet.max_row

# 遍历 A 列的每一行
for row in range(1, max_row + 1):
    # 获取 A 列当前行的单元格值，即 sheet 名
    sheet_name = source_sheet.cell(row=row, column=1).value
    tname=source_sheet.cell(row=row, column=2).value

    # 检查单元格值是否为空
    if sheet_name is not None:
        # 检查该名称的 sheet 是否存在于工作簿中
        name=''
        if sheet_name in workbook.sheetnames:
            name=sheet_name
        elif sheet_name.lower() in workbook.sheetnames:
            name=sheet_name.lower()
        if name!='': 

            target_sheet = workbook[name]
            chname = target_sheet.cell(row=1, column=2).value
            print(chname)
            # 获取目标 sheet 的最大行数
            target_max_row = target_sheet.max_row

            # 找到源工作表的最后一行
            last_row = source_sheet.max_row

            # 遍历目标 sheet 的 D 列至 F 列的每一行
            for target_row in range(3, target_max_row + 1):

                for col in range(1, 7):  # D 列到 F 列对应的列索引为 4 到 6
                    if col==1:
                        source_sheet.cell(row=last_row + 1, column=col).value = tname
                    else:
                        value = target_sheet.cell(row=target_row, column=col).value
                        print(value)
                        if isinstance(value, str):
                            value = value.upper()
                        source_sheet.cell(row=last_row + 1, column=col).value = value
                value = target_sheet.cell(row=target_row, column=4).value    
                if isinstance(value, str):
                    pattern = r'^(.*?)\((.*?)\)'
                    match = re.match(pattern,value)
                    pattern2 = r'(\w+)\（([^）]+)）'
                    match2 = re.match(pattern2,value)
                    if match:
                        text_before = match.group(1).strip()
                        if isinstance(text_before, str):
                                text_before = text_before.upper()
                        text_inside = match.group(2).strip()
                        source_sheet.cell(row=last_row + 1, column=8).value = text_before
                        source_sheet.cell(row=last_row + 1, column=9).value = text_inside
                    elif match2:
                        text_before = match2.group(1).strip()
                        if isinstance(text_before, str):
                                text_before = text_before.upper()
                        text_inside = match2.group(2).strip()
                        source_sheet.cell(row=last_row + 1, column=8).value = text_before
                        source_sheet.cell(row=last_row + 1, column=9).value = text_inside
                    else:
                        source_sheet.cell(row=last_row + 1, column=8).value = value.upper()
                last_row += 1

# 保存修改后的工作簿
workbook.save('your_excel_file_modified.xlsx')

# 关闭工作簿
workbook.close()
```
